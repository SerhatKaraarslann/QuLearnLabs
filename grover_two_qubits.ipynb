Grover's Algorithm with 2 Qubits


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
```

```python
qc = QuantumCircuit(2)

qc.h(0)
qc.h(1)

#tag 11
#qc.cz(0,1)

#tag 01
#qc.cz(0,1)
#qc.x(1)

#tag 10
#qc.cz(0,1)
#qc.x(0)

#tag 00
qc.cz(0,1)
qc.x(0)
qc.x(1)

state = Statevector(qc)
print("State of quantum circuit :",state)

qc.draw("mpl")
```

```python
from qiskit.primitives import StatevectorSampler
```

```python
qc = QuantumCircuit(2)

qc.h(0)
qc.h(1)

qc.cz(0,1)
qc.x(0)
qc.x(1)

qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)

qc.cz(0,1)


qc.x(0)
qc.x(1)

qc.h(0)
qc.h(1)

qc.measure_all()

sampler = StatevectorSampler()

job = sampler.run([qc],shots = 1)

result = job.result()[0]

state = result.data.meas.get_bitstrings()

print(state)

qc.draw("mpl")
```
